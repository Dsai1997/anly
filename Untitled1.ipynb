```python
from IPython.display import HTML

HTML('''
<iframe width="600" height="520" style="border:2px solid #ccc;" srcdoc="
<!DOCTYPE html>
<html>
<head>
<style>
  body {
    margin: 0;
    font-family: 'Segoe UI', sans-serif;
    background: url('https://images.unsplash.com/photo-1507525428034-b723cf961d3e') no-repeat center center fixed;
    background-size: cover;
    display: flex;
    flex-direction: column;
    align-items: center;
  }
  canvas {
    background: #ffffffcc;
    display: block;
    border: 2px solid #aaa;
    margin-top: 20px;
    box-shadow: 0 4px 10px rgba(0,0,0,0.3);
  }
  .controls {
    margin-top: 20px;
    background: rgba(255,255,255,0.8);
    padding: 10px 20px;
    border-radius: 12px;
    box-shadow: 0 4px 8px rgba(0,0,0,0.2);
  }
  #restartBtn {
    border-radius: 10px;
    background-color: #d63031;
    padding: 10px 25px;
    font-size: 18px;
    display: none;
    color: white;
    border: none;
    cursor: pointer;
  }
</style>
</head>
<body>
<canvas id='gameCanvas' width='600' height='400'></canvas>

<div class='controls'>
  <button class='btn' id='restartBtn'>ðŸ”„ Restart</button>
</div>

<script>
let canvas = document.getElementById('gameCanvas');
let ctx = canvas.getContext('2d');

let ball, paddle, moveLeft, moveRight, gameOver;

function initGame() {
  ball = {
    x: canvas.width / 2,
    y: canvas.height - 30,
    dx: 3,
    dy: -3,
    radius: 10
  };
  paddle = {
    height: 10,
    width: 100,
    x: (canvas.width - 100) / 2,
    speed: 6
  };
  moveLeft = false;
  moveRight = false;
  gameOver = false;
  document.getElementById('restartBtn').style.display = 'none';
  draw();
}

// Keyboard only
document.addEventListener('keydown', function(e) {
  if (e.key === 'ArrowRight') moveRight = true;
  if (e.key === 'ArrowLeft') moveLeft = true;
});
document.addEventListener('keyup', function(e) {
  if (e.key === 'ArrowRight') moveRight = false;
  if (e.key === 'ArrowLeft') moveLeft = false;
});

// Restart button
document.getElementById('restartBtn').onclick = initGame;

function drawBall() {
  ctx.beginPath();
  ctx.arc(ball.x, ball.y, ball.radius, 0, Math.PI*2);
  ctx.fillStyle = '#FF4D4D';
  ctx.fill();
  ctx.closePath();
}

function drawPaddle() {
  ctx.beginPath();
  ctx.rect(paddle.x, canvas.height - paddle.height - 10, paddle.width, paddle.height);
  ctx.fillStyle = '#339933';
  ctx.fill();
  ctx.closePath();
}

function draw() {
  if (gameOver) return;

  ctx.clearRect(0, 0, canvas.width, canvas.height);
  drawBall();
  drawPaddle();

  if(ball.x + ball.dx > canvas.width - ball.radius || ball.x + ball.dx < ball.radius) {
    ball.dx = -ball.dx;
  }

  if(ball.y + ball.dy < ball.radius) {
    ball.dy = -ball.dy;
  } else if(ball.y + ball.dy > canvas.height - ball.radius - 10) {
    if(ball.x > paddle.x && ball.x < paddle.x + paddle.width) {
      ball.dy = -ball.dy;
    } else if(ball.y + ball.dy > canvas.height - ball.radius) {
      gameOver = true;
      document.getElementById('restartBtn').style.display = 'inline-block';
      return;
    }
  }

  if (moveRight && paddle.x < canvas.width - paddle.width) {
    paddle.x += paddle.speed;
  } else if (moveLeft && paddle.x > 0) {
    paddle.x -= paddle.speed;
  }

  ball.x += ball.dx;
  ball.y += ball.dy;

  requestAnimationFrame(draw);
}

initGame();
</script>
</body>
</html>
"></iframe>
''')
```
